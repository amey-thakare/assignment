```python
order=[1200,2500,800,1750,3000]
cnt=0
for order_amount in order:
    if order_amount>=2000:
        discount=0.15
        
    elif order_amount<2000 and order_amount>=1500:
        discount=0.1
        
    elif order_amount<1500 and order_amount>=1000:
        discount=0.07
        
    else:
        discount=0
    if discount>0 :
        cnt=cnt+1
    subtotal=order_amount-(order_amount)*discount
    
    print("Subtotal:",subtotal)
    print("tax:" ,subtotal*0.05)
    print("Final total:",subtotal+subtotal*0.05)
    print(cnt," people got the discount")
```
